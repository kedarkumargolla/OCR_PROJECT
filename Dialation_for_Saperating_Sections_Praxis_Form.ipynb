```python
import cv2
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

#Load the image in black and white (0 - b/w, 1 - color).
img = cv2.imread('P47A27UB4Kb2A2RKt5dPY4t8.jpg', 0)

#Get the height and width of the image.
h, w = img.shape[:2]

#Invert the image to be white on black for compatibility with findContours function.
imgray = 255 - img
#Binarize the image and call it thresh.
#ret, thresh = cv2.threshold(imgray, 127, 255, cv2.THRESH_BINARY)

#kernel = np.ones((5,2),np.uint8)
#imgray = cv2.dilate(imgray,kernel,iterations = 1)

kernel = np.ones((2,5),np.uint8)
imgray = cv2.dilate(imgray,kernel,iterations = 1)

kernel = np.ones((5,2),np.uint8)
imgray = cv2.dilate(imgray,kernel,iterations = 1)

thresh = cv2.adaptiveThreshold(imgray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
#Find all the contours in thresh. In your case the 3 and the additional strike
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
#Calculate bounding rectangles for each contour.
rects = [cv2.boundingRect(cnt) for cnt in contours]
```

```python
rects
```

```python
bigger_contours=[]
for cont in rects:
    if cont[0]<130  and cont[0]>100 and cont[2]>450 and cont[3]>100:
       bigger_contours.append(cont) 
```

```python
from operator import itemgetter
k=sorted(bigger_contours,key=itemgetter(1))
len(k)
k
```

```python
initial_list=k.copy()
for i in range(len(k)-1):
    try:
        
        for j in range(i+1,len(k)):
            if abs(k[i][1] -k[j][1]) <40 and abs(k[i][3]-k[j][3]) < 50 :
                if k[i][2]>k[j][2]:
                    value=k[j]
                    initial_list.remove(value)
                else :
                
                    value=k[i]
                    initial_list.remove(value)
    except:
        continue
        
```
